# Assignment I

```python
#1 Creating 4 list with name EmpID, Fname, Lname, Designation, Basic_Salary with minimum 5 record in each list
#importing libraries
import pandas as pd #Data preprocessing 
import numpy as np #Numerical computattion
EmpID = [1, 2, 3, 4, 5, 6]
Fname = ['Riks', 'Bob', 'Joe', 'Lily', 'Ben']
Lname = ['Singh', 'Weryn', ' ' , 'Oliver', ' ']
Designation = ['Jr. Developer', 'Team Leader', 'Sr. Developer', 'Tester', 'Manager']
Basic_Salary = [18000, 25000, 45000, 20000, 54000]
```

```python
#2. Create table of above list with the name "Empdata"
Empdata = zip(EmpID, Fname, Lname, Designation, Basic_Salary)
```

```python
#3. Convert Empdata as dataframe
Empdata = pd.DataFrame(data=Empdata, columns=["EmpID", "Fname", "Lname", "Designation", "Basic_Salary"])
Empdata
```

```python
#4. Convert Empdata to csv file and read contents of same csv file
Empdata.to_csv('Empdata csv', index = False)
Empdata_csv = pd.read_csv('Empdata csv')
Empdata_csv
```

```python
#5. Convert Empdata to excel file and read contents of same excel file
writer = pd.ExcelWriter('Empdata.xlsx', engine='xlsxwriter')
Empdata.to_excel(writer, sheet_name='sheet1')
writer.save()
reading_xlsx = pd.read_excel("Empdata.xlsx")
reading_xlsx
```

```python
#6. Convert Empdata to sqlite, create database with name Employee.db and read contents of file.
import sqlite3
con = sqlite3.connect('Employee.db')
cur = con.cursor()
cur.execute('Create Table EmployeeData(EmpID number, Fname Text, Lname Text, Designation Text, Basic_Salary Number)')
con.commit()
Empdata.to_sql('EmployeeData', con, if_exists='replace', index=False)
Empdata
```

```python
#7. Retrieve all the records of Employee.db whose salary in the range 20000-35000
cur.execute('''
SELECT * from EmployeeData where Basic_Salary> 20000 and Basic_Salary< 35000
''')
df = pd.DataFrame(cur.fetchall(), columns=['EmpID', 'Fname', 'Lname', 'Designation', 'Basic_Salary'])
df
```
